#### Compute the cosine similarity between two words

cosine similarity = $\frac{|a.b|}{|a||b|}$

```python
cosine_similarity = model.wv.similarity("trees", "user")
cosine_similarity
```

**Manual computation of cosine similarity**

```python
vector_for_tree = model.wv["trees"]
vector_for_user = model.wv["user"]

magnitude_trees = np.sqrt(vector_for_tree.dot(vector_for_tree))
magnitude_user = np.sqrt(vector_for_user.dot(vector_for_user))

magnitude_dot_product = vector_for_tree.dot(vector_for_user)
magnitude_dot_product/(magnitude_user*magnitude_trees)
```

#### Compute the distance between two words

cosine distance = 1 - cosine similarity

```python
distance = model.wv.distance('trees', 'user')

distance
```

```python
cosine_similarity + distance
```

```python
model = Word2Vec(corpus)


model.wv.index_to_key # get the words in the vocabulary
```

```python
model.wv.most_similar("countries")
```
